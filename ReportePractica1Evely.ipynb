# Introducción

<p style="text-align: justify;"> 
    En el siguiente reporte se analiza el caso de estudio que permitirá evaluar las condiciones de reinserción social dentro del CERESO "Apodaca" y dentro del Centro Preventivo de Reinserción Topo Chico. Los datos que se usan para este estudio provienen de encuestas realizadas a mujeres y hombres por separado, que se encuentran privados de libertad dentro del Centro de Reinserción. Con este cuestionario se pretende evaluar las condiciones de vida, económicas, educativas y familiares de estas personas.
</p>


Se conforman diferentes preguntas con respecto a estos datos, que permitirán evaluar las condiciones de los internos en estos centros.
1. ¿El nivel de escolaridad tendrá valor significativo en la necesidad de cometer un delito? Es decir, comprobar si el tener nivel escolar bajo pudiera afectar significativamente la necesidad de cometer un delito.
2. ¿Que puede provocar que la familia del interno se vea afectada en la sociedad por el encarcelamiento de un familiar?
3. ¿Tendrá sentido afirmar que al ser una persona joven (entre 15 y 26), al entrar en el CERESO es muy probable que comience a tener adicciones al cigarro, alcohol, solventes u otras drogas?


# Desarrollo

<p style="text-align: justify;"> 
Se cuenta con dos encuestas, una aplicada a los hombres y otra a las mujeres, con alrededor de 287 preguntas cada una y una sesión de autoevaluación dele estado de ánimo del encuestado. Las dos encuestas son muy parecidas, solo hacen diferencias en algunas preguntas referidas a las condiciones que tienen las mujeres en el penitenciario para tener a sus bebés, en caso que hallan dado a luz a su bebé estando bajo cumplimiento de condena, dado que viven con ellas sus primeros 2 años de vida. 
<p style="text-align: justify;"> 
Las preguntas de los cuestionarios giran sobre temas como: la infraestructura y los servicios de la penitenciaria, la situación familiar de los reclusos, si se realiza trabajo social con ellos, si los atienden clínicamente en caso de adicciones así como atención psicológica, sobre los servicios educativos y de capacitación, contiene temas de cultura y deporte, del ámbito laboral antes de ingresar al penitenciario y dentro del mismo, si cuentan o no con servicios de asesoría penal y jurídica, abordan el caso de la reinserción social de los reclusos y se tienen una sesión de preguntas referentes al estado de anímico del recluso. 
<p style="text-align: justify;"> 
Se conoce que estas encuestas fueron respondidas por aproximadamente 313 hombre y 173 mujeres que se encuentran privados de libertad en CERESO en Apodaca. Fueron realizadas en papel y as respuestas se digitalizaron mediante formato CSV. Estos datos ya se les apicó una limpieza parcial anteriormente, por tanto se procederá a revisar y limpiar nuevamente en caso de ser necesario. 
<p style="text-align: justify;"> 
Se utilizará herramientas de Bash para el manejo de estos archivos. Se comienza la nueva limpieza de datos haciendo copias de los archivos que se van a utilizar, en este caso, backup-male.csv para las encuestas de los hombre, backup-female.csv para las de mujeres y backup-uniform.csv para la unión de los datos de hombres y mujeres. Con estas copias se garantiza no dañar los datos originales en caso de cometer errores.


-----

`cp uniform.csv backup-uniform.csv
cp female.csv backup-female.csv
cp male.csv backup-male.csv`

<p style="text-align: justify;"> 
Luego para comenzar el análisis de los datos, se procede a separar solamente las columnas que estan involucradas con el análisis en cuestión, que dará respuesta a las preguntas planteadas en el inicio del reporte. Para esto se utiliza el comando **awk** y se seleccionan las columnas involucradas, en este caso son las columnas 1, 6, 12, 29 del fichero **backup-male.csv** y se almacenan en el otro fichero **pregunta1male.csv**.

`awk -F',' '{print $1","$6","$12","$29}' backup-male.csv > pregunta1male.csv`

<p style="text-align: justify;"> 
Siguiendo con el uso del comando **awk** y las herramientas **sort** y **uniq** se puede ordenar las columnas deseadas y unificar por respuestas, para facilitar el analisis de los datos. En este ejemplo se trabajo con los datos de los hombres de la primera pregunta y se ordenó y unificó por el nivel de escolaridad. Del mismo modo se realizó este trabajo para las columnas edad y si es primera vez que está interno o no.
